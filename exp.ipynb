```python
from edgar import Company, XBRL
from edgar.xbrl.stitching import to_pandas

def stitch_company(ticker, form="10-Q", count=4, stmt="IncomeStatement", period_type=None, max_periods=4):
    company = Company(ticker)
    filings = company.get_filings(form=form)
    xbrl_list = [XBRL.from_filing(f) for f in filings]
    stitched = XBRL.stitch_statements(
        xbrl_list,
        statement_type=stmt,
        period_type=period_type,  # e.g. StatementStitcher.PeriodType.RECENT_PERIODS
        max_periods=max_periods,
        standard=True
    )
    df = to_pandas(stitched)
    # Add a company prefix to columns so they don't collide when merging
    df = df.add_prefix(f"{ticker}_")
    return df

aapl_df = stitch_company("AAPL", count=4)
msft_df = stitch_company("MSFT", count=4)

# Align periods (index) and join
import pandas as pd
combined = aapl_df.join(msft_df, how="outer")
```

```python
from edgar.xbrl.standardization.core import MappingStore, initialize_default_mappings

# Provide the path to the concept_mappings.json file
# If the file doesn't exist, the `initialize_default_mappings()` function will create it (not read_only mode)
mapping_store = initialize_default_mappings()
```

```python
mapping_store
```

```python
from edgar.xbrl.standardization.core import ConceptMapper

# Initialize the ConceptMapper with the MappingStore
concept_mapper = ConceptMapper(mapping_store)

# Example company label from a statement line item
company_concept = "us-gaap_Revenue"  # XBRL tag or label
concept_label = "Total Revenue for FY2022"  # Readable label
context = {
    "statement_type": "IncomeStatement",  # Expected statement type
    # Add additional logic (optional)
}

# Get the standard concept for the company-specific label
standard_concept = concept_mapper.map_concept(company_concept, concept_label, context)

print(f"Standard Concept for '{company_concept}': {standard_concept}")
```

```python
from edgar.xbrl.standardization.core import standardize_statement

# Example: Statement data from multiple companies
# Each item should include the statement_type, concept, label, etc.
company_statement_data = [
    {"concept": "us-gaap_Revenue", "label": "Total Revenue", "statement_type": "IncomeStatement"},
    {"concept": "us-tsla_Sales", "label": "Sales (Tesla)", "statement_type": "IncomeStatement"},
    {"concept": "us-aapl_OpIncome", "label": "Operating Income", "statement_type": "IncomeStatement"}
    # More records...
]

# Automatically map the data's company-specific concepts to standardized ones
standardized_data = standardize_statement(company_statement_data, concept_mapper)

# Output the standardized statement data
for item in standardized_data:
    print(item)
```

```python
# Add mappings for a company-specific concept to a standard concept
company_specific_concept = "us-aapl_TotalRevenue"
standard_concept = "Revenue"

mapping_store.add(company_specific_concept, standard_concept)
print(f"Mapping added: {company_specific_concept} -> {standard_concept}")
```
